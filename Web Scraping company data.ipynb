```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python


URL = 'https://www.ambitionbox.com/list-of-companies?page=1'
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}

webpage = requests.get(URL,headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
company=soup.find_all('div',class_= 'company-content-wrapper')
```

```python
len(company)
```

```python
name = []
rating =[]
reviews=[]
ctype=[]
hq=[]
old=[]
employee=[]


for i in company:
    name.append(i.find('h2').text.strip())
    rating.append(i.find('p',class_='rating').text.strip())
    reviews.append(i.find('a',class_='review-count').text.strip()[1:-1])
    ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
    hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
    old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
    try:
        employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
    except:
        employee.append(np.nan)
    
d= {'name':name,'rating':rating, 'reviews':reviews,'ctype':ctype,'hq':hq,'old':old,'employee':employee }   
    
df = pd.DataFrame(d)   
```

```python
final=pd.DataFrame()
for j in range(1,11):
    URL = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}

    webpage = requests.get(URL,headers=headers).text
    
    soup = BeautifulSoup(webpage,'lxml')
    
    company=soup.find_all('div',class_= 'company-content-wrapper')
    
    name = []
    rating =[]
    reviews=[]
    ctype=[]
    hq=[]
    old=[]
    employee=[]


    for i in company:
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',class_='rating').text.strip())
        reviews.append(i.find('a',class_='review-count').text.strip()[1:-1])
        ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
        try:
            hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
        except:
            hq.append(np.nan)    
        try:
            old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
        try:
            employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
        except:
            employee.append(np.nan)

    d= {'name':name,'rating':rating, 'reviews':reviews,'ctype':ctype,'hq':hq,'old':old,'employee':employee }   

    df = pd.DataFrame(d)
    
    final=final.append(df,ignore_index=True)
    
```
